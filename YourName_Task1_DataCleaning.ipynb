# Task 1: Data Cleaning and Preprocessing

```
import pandas as pd

df = pd.read_csv('spotify_history.csv')

print(df.head())
print(df.info())
print(df.describe(include='all'))
```

```
# Missing values
print(df.isnull().sum())

# Duplicates
print(df.duplicated().sum())
```

```
# Cleaning
df = df.drop_duplicates()
df['reason_start'] = df['reason_start'].fillna(df['reason_start'].mode()[0])
df['reason_end'] = df['reason_end'].fillna(df['reason_end'].mode()[0])
df['ts'] = pd.to_datetime(df['ts'])

# Standardization
for col in ['track_name','artist_name','album_name','platform','reason_start','reason_end']:
    df[col] = df[col].astype(str).str.strip()

df.to_csv('spotify_history_cleaned.csv', index=False)
```
